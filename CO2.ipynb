# Which countries are the largest contributors of CO2?

```python
df.country.unique()
```

```python
l=['Upper-middle-income countries','World','North America (excl. USA)',
'Low-income countries', 'Lower-middle-income countries','High-income countries','European Union (28)'
,'European Union (27)','Europe (excl. EU-28)','Europe (excl. EU-27)','Asia (excl. China and India)',
'Asia','Africa','North America','South America','Lower-middle-income countries','Europe','International transport']
df['is region']=df['country'].apply(lambda x:True if (x[-5:-1]=='(GCP' or x in l) else False)
df.loc[df.country.isin(['Asia']),['country','is region']]
```

```python
countries_co2=df.loc[df['is region']==False,['country','co2']]
count=countries_co2['country'].value_counts()
fig,ax=plt.subplots(figsize=(40,10))
ax.bar(count.index,count.values)
ax.set_xticklabels(count.index,rotation=90)
plt.show()
```

```python
sum_co2_country=countries_co2.groupby('country').co2.sum().sort_values(ascending=False).head(10)
fig,ax=plt.subplots(figsize=(40,10))
ax.bar(sum_co2_country.index,sum_co2_country.values)
ax.set_xlabel("Countries",fontsize=20,fontweight='bold')
ax.set_ylabel("summetion of CO2 contribution over years",fontsize=20,fontweight='bold')
ax.set_title("The top 10 countries have CO2 contributions",fontsize=20,fontweight='bold')
plt.xticks(rotation=90,fontsize=15,fontweight='bold')
plt.yticks(fontsize=15,fontweight='bold')
plt.show()
```

#Create a line chart showing the % of total share of CO2 by year for the top 10 countries using your TOP N parameter (remove any null value countries)

```python
countries_co2=df.dropna(subset=['co2'])
countries_co2
```

```python
countries_co2=countries_co2.reset_index()
countries_co2=countries_co2.drop(columns=['index'])
countries_co2
```

```python
countries_co2['co2'].isnull().sum()
```

```python
top_10_countries=countries_co2.loc[countries_co2['country'].isin(sum_co2_country.index)]
countries_co2_per_year=top_10_countries.groupby(['year','country']).co2.sum()
total_co2=top_10_countries.groupby('year').co2.sum()
countries_co2_per_year=(countries_co2_per_year/total_co2)*100
countries_co2_per_year=countries_co2_per_year.reset_index()

fig2,ax2=plt.subplots(figsize=(30,10))
for country in countries_co2_per_year['country'].unique():
    subset = countries_co2_per_year[countries_co2_per_year['country'] == country]
    ax2.plot(subset['year'], subset['co2'], label=country)

ax2.set_xlabel("years",fontsize=20,fontweight='bold')
ax2.set_ylabel("CO2 Percentage",fontsize=20,fontweight='bold')
ax2.set_title("The Percentage of CO2 to top 10 countries contributions",fontsize=20,fontweight='bold')
plt.xticks(fontsize=20,fontweight='bold')
plt.yticks(fontsize=20,fontweight='bold')
plt.legend(fontsize=15)
plt.show()
```

#Create a map at country level using Co2 Per Capita for the year 2021 only (fix any country/region errors, remove null value countries)

```python
# country / region error solved by deleted iso
df['country'].unique()
```

```python
country_co2_per_capita=df.dropna(subset=['co2_per_capita'])
country_co2_per_capita.co2_per_capita.isnull().sum()
```

```python
country_co2_per_capita_group=country_co2_per_capita.groupby('country').co2_per_capita.sum().sort_values(ascending=False)
country_co2_per_capita_group=country_co2_per_capita_group.reset_index()
country_co2_per_capita_group
```

```python
country_co2_per_capita_group['co2_per_capita'].describe()
```

```python
labels = ['Very Low', 'Low', 'Medium', 'High', 'Very High']
country_co2_per_capita_group['quantile_bin'] = pd.qcut(country_co2_per_capita_group['co2_per_capita'], q=5, labels=labels)
```

```python
country_co2_per_capita_group
```

```python
import plotly.express as px
```

```python
color_map = {
    'Very Low': '#ffffb2',
    'Low': '#fecc5c',
    'Medium': '#fd8d3c',
    'High': '#f03b20',
    'Very High': '#bd0026'
}

fig = px.choropleth(
    country_co2_per_capita_group,
    locations="country",
    locationmode="country names",
    color="quantile_bin",
    color_discrete_map=color_map,
    title="COâ‚‚ Per Capita by Country"
)

fig.show()
```
